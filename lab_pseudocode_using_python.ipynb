<!-- Add your pseudocode here -->

```python
"""

BEGIN
    SET N = 10
    SET sum = 0
    FOR i FROM 1 TO N
        sum = sum + i
    END FOR
    PRINT sum
END

"""
```

<!-- Add your pseudocode here -->

### Bonus Exercise 8: Count Words in a String

Write pseudocode that counts the words in a given string and prints the result.

```python
"""

BEGIN

    DEFINE FUNCTION word_counter(text):

      READ text

      SET list_of_words = text.split(" ")
      PRINT list_of_words

      FOR words in list_of_words:
        count = len(list_of_words)
      PRINT(count)

END

"""
```

### Bonus Exercise 9: Find Factorial of a Number

Write pseudocode that calculates the factorial of a given positive integer.

<!-- Add your pseudocode here -->

```python
"""

BEGIN

    DEFINE FUNCTION calc_factorial(number)

    READ number
    SET result = 1

    FOR n in range(1, number + 1)
      SET result *= n

    PRINT(result)

END

"""
```

```python
def calc_factorial(number):
  result = 1

  for i in range(1, number+1):
    result *= i

  print(result)
```

```python
calc_factorial(0)
```

### Bonus Exercise 10: FizzBuzz Problem

Write pseudocode for the FizzBuzz problem where you print numbers from 1 to N but replace multiples of three with "Fizz" and multiples of five with "Buzz".

<!-- Add your pseudocode here -->

```python
"""

BEGIN

    DEFINE FUNCTION fizzebuuble(n)

    READ n
      FOR i in range(1, number+1):
        IF i % 3 == 0:
          PRINT("Fizze")
        ELSE IF i % 5 == 0:
          PRINT("Buzze")

      PRINT(i)

END

"""
```

```python
def fizzebuuble(n):
  for i in range(1, n):
    if i % 3 == 0:
      print("FIZZE")
    elif i % 5 == 0:
      print("BUZZE")
    else:
      print(i)
```

```python
fizzebuuble(100)
```
